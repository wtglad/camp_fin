```python
# Scale data for clustering
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_df = scaler.fit_transform(cp2008_concf_known[feature_cols])
```

```python
scaled_df
```

```python
pd.DataFrame(scaled_df, columns=feature_cols)
```

# Run k-means clustering

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_jobs=-1)
```

```python
clusters = km.fit_predict(scaled_df)
```

```python
cp2008_concf_known['cluster'] = clusters
```

```python
for i in feature_cols:
    sns.distplot(cp2008_concf_known[i])
    plt.show()
```
